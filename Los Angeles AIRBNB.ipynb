```python
# Checking Null Values using Heatmap
plt.figure(figsize=(15,8))
a=labnb.isnull()
sns.heatmap(a,cbar_kws={'label':'Missing Data'})
plt.title('Heatmap for Null values')
plt.show()
```

```python
# Checking null values in each column
labnb.isnull().sum().sort_values(ascending=False)
```

```python
# Pecentage of missing values in each column
labnb.isnull().sum().sort_values(ascending=False)/labnb.shape[0]
```

```python
# Handling Missing Values
labnb.reviews_per_month=labnb.reviews_per_month.fillna(labnb.reviews_per_month.mean())
```

```python
for i in labnb.columns:
    labnb[i]=labnb[i].fillna(labnb[i].value_counts().idxmax())
```

```python
# Finding Skewness And Kurtosis
print('Skewness of Price:',labnb.price.skew())
print('Kurtosis of Price:',labnb.price.kurt())
```

```python
# Finding Frequency Counts of Categorical Variable
a=['neighbourhood_group','neighbourhood','room_type','last_review_year','last_review_month']
for i in a:
    freq=labnb[i].value_counts()
    print(freq)
```

```python
# Top10 Neighbourhood
plt.figure(figsize=(10,6))
a=labnb.neighbourhood.value_counts().nlargest(10).plot(kind='bar',color='red')
for i in a.containers:
    a.bar_label(i,fontsize=8)    
plt.xticks(rotation=45)
plt.title('Top10 Frequency Counts of Neighbourhood')
plt.ylabel('Frequency',fontsize=12)
plt.xlabel("Neighbourhood",fontsize=12);
```

```python
# Top10 Neighbourhood
plt.figure(figsize=(10,6))
a=labnb.neighbourhood.value_counts().nsmallest(10).plot(kind='bar',color='violet')
for i in a.containers:
    a.bar_label(i,fontsize=8)    
plt.xticks(rotation=45)
plt.title('Bottom10 Frequency Counts of Neighbourhood')
plt.ylabel('Frequency',fontsize=12)
plt.xlabel("Neighbourhood",fontsize=12);
```

```python
# Cross tabulations - Frequency Table of 2 Categorical Variables
# Cross Tabulation of Neighbourhood & Neighbourhood_group
a=pd.crosstab(labnb.neighbourhood,labnb.neighbourhood_group).nlargest(10,columns="City of Los Angeles").plot(
    kind='bar',figsize=(10,6))
for i in a.containers:
    a.bar_label(i,fontsize=10)
plt.title('Cross Tabulation Between Neighbourhood and Neighbourhood_group of City of Los Angeles Type')
plt.xticks(rotation=45)
plt.ylabel('Neighbourhood_group',fontsize=10)
plt.xlabel('Neighbourhood',fontsize=10);
```

```python
# Cross Tabulation of Neighbourhood & Room_type of Entire home/apt
a=pd.crosstab(labnb.neighbourhood,
            labnb.room_type).nlargest(10,columns="Entire home/apt").plot.bar(stacked=True,figsize=(10,6))
for i in a.containers:
    a.bar_label(i,fontsize=8,label_type="center")
plt.xticks(rotation=45)
plt.title('Cross Tabulation Between Neighbourhood and Room_type of Entire home/apt')
plt.ylabel('Room_type',fontsize=10)
plt.xlabel('Neighbourhood',fontsize=10);
```

```python
a=np.round(labnb.minimum_nights.groupby(labnb.neighbourhood_group).mean()).plot(kind='bar',figsize=(10,6),color='cyan')
for i in a.containers:
    a.bar_label(i,fontsize=10)
    plt.xticks(rotation=0)
plt.title('Average of Minimum_nights for Neighbourhood_group')
plt.ylabel('Average Minimum_Nights',fontsize=10)
plt.xlabel('Neighbourhood_group',fontsize=10);
```
